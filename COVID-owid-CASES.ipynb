```python
df_cases = df[["date","new_cases_smoothed_per_million"]]
```

```python
df_france = df_cases.loc[df['location'] == "France" ]
df_germany = df_cases.loc[df['location'] == "Germany" ]
df_italy = df_cases.loc[df['location'] == "Italy" ]
df_japan = df_cases.loc[df['location'] == "Japan" ]
df_uk = df_cases.loc[df['location'] == "United Kingdom" ]
df_usa = df_cases.loc[df['location'] == "United States" ]
df_canada = df_cases.loc[df['location'] == "Canada" ]
df_russia = df_cases.loc[df['location'] == "Russia" ]
df_spain = df_cases.loc[df['location'] == "Spain" ]
```

```python
#result_merge = pd.merge(df_france, df_germany, df_italy, df_japan, df_uk, df_usa, df_canada, df_russia, how ="inner", on = "date")
result_merge_fr_gr = pd.merge(df_france, df_germany, how ="inner", on = "date")
```

```python
result_merge_it = pd.merge(result_merge_fr_gr, df_italy, how ="inner", on = "date")
result_merge_jp = pd.merge(result_merge_it, df_japan, how ="inner", on = "date")
result_merge_uk = pd.merge(result_merge_jp, df_uk, how ="inner", on = "date")
result_merge_usa = pd.merge(result_merge_uk, df_usa, how ="inner", on = "date")
result_merge_ca = pd.merge(result_merge_usa, df_canada, how ="inner", on = "date")
result_merge_ru = pd.merge(result_merge_ca, df_russia, how ="inner", on = "date")
```

```python
result_merge_ru
```

```python
result_total = result_merge_ru
result_total.columns = ['date', 'France', 'Germany', 'Italy', 'Japan', 'UK', 'USA', 'Canada', 'Russia']
```

```python
result_total
```

```python
plt.figure();

result_total.plot(figsize=(20, 10));
```

```python
plt.close("all")
```

```python
df_today_g8 = result_total[result_total["date"] == str(yesterday)]
index_num_g8 = df_today_g8.index.values.astype(int)[0]
index_num_g8
```

```python
four_months_ago_g8 = index_num_g8 - 120
```

```python
recent_df_g8 = result_total.loc[four_months_ago_g8:index_num_g8]
recent_df_g8
```

```python
plt.figure();

recent_df_g8.plot(figsize=(20, 10));
```

```python
plt.close("all")
```

```python
result_total_spain = pd.merge(result_total, df_spain, how ="inner", on = "date")
```

```python
recent_df_g8_spain = result_total_spain.loc[four_months_ago_g8:index_num_g8]
recent_df_g8_spain.columns = ['date', 'France', 'Germany', 'Italy', 'Japan', 'UK', 'USA', 'Canada', 'Russia', 'Spain']
recent_df_g8_spain
```

```python
plt.figure();

recent_df_g8_spain.plot(figsize=(20, 10));
```

```python
plt.close("all")
```
